```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set()
sns.set_theme(style="whitegrid", context="poster")
```

```python
pos = np.array([[967,   619, 1337,  754, 1323, ],
                [1019,   450, 1419,  857, 1255, ],
                [1124,   482, 1205, 1022, 1167, ],
                [1923,   92,  372, 2329,  284, ],
                [2103,  779,  729,  778,  611, ]])
row_sums = pos.sum(axis=1)
dota = pos / row_sums[:, np.newaxis]
# np.round(pick_positions_normalized, 3)

lol = np.array([[0.1154, 0.2196, 0.3274, 0.2394, 0.0982],
                [0.194, 0.1886, 0.2326, 0.2818, 0.103],
                [0.1202, 0.2316, 0.264, 0.2444, 0.1398],
                [0.0812, 0.1796, 0.2488, 0.333, 0.1574],
                [0.034, 0.2474, 0.1004, 0.4872, 0.131]])
# print(np.round(lol, 3))
```

```python
df_dota = pd.DataFrame(dota, columns=['Position 1', 'Position 2', 'Position 3', 'Position 4', 'Position 5'], index=['Pick 1', 'Pick 2', 'Pick 3', 'Pick 4', 'Pick 5'])
df_lol = pd.DataFrame(lol, columns=['Bottom', 'Mid', 'Top', 'Jungle', 'Utility'], index=['Pick 1', 'Pick 2', 'Pick 3', 'Pick 4', 'Pick 5'])
```

```python
sns.set_palette("tab10")

ax = df_dota.plot(kind='bar', figsize=(10, 5),rot=0, ylim=(0, 0.5))
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5)) #here is the magic
```
