```python
import numpy as np
import pickle
import glob
from scipy import sparse
from collections import Counter


hero_roles = np.zeros((1,605), dtype=int)

for file in glob.iglob('../data/vl_train_data_full/*.pkl'):
    with open(file, 'rb') as f:
        data = pickle.load(f)
    
    xs = data['x']
    counts = np.abs(xs).sum(axis=0).astype(int)
    
    hero_roles += counts


print(hero_roles)
```

```python
# 

hero_roles = hero_roles.reshape(-1,5)
z = hero_roles.sum(axis=1)
pi = hero_roles / z[:,None]

for i in range(121):
    c = pi[i]
    # print(c)


np.save('../a0/hero_roles.npy', hero_roles)
```
